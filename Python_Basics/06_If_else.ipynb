CONDITIONAL STATEMENTS :-

```python
x = 5
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but not more than 10")
elif x == 5:
    print("x is exactly 5")
else:
    print("x is less than 5")
```

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

FUNCTION IN PYTHON :-   "A Function is a reusable block of code which perform a specific task "

```python
def add(x,y):
    return x+y
```

```python
add(4,5)
```

```python
def avg(a,b):
    return (a+b)/2
```

```python
avg(3,3)
```

```python
def avg(a,b=9):        #default argument and it shoud be in last
    return (a+b)/2
```

```python
avg(9)
```
