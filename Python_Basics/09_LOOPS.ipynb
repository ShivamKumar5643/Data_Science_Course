LOOP CONTROL STATEMENTS :- PASS,BREAK,CONTINUE

```python
for i in range(7):
    pass         # do nothing , also use in function and loop


print("hello")
```

```python
l =["apple","banana","tomato"]
for item in l:
    if (item=="banana"):
        break      #Stop this loop right here
    print(item)
```

```python
l = ["apple","banana","tomato"]
for item in l:
    if(item == "banana"):
        continue     #Skip this particular iteration of the loop
    print(item)
```
