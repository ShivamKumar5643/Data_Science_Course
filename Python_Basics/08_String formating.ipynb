String formating:-

```python
name = "Shivam"
age = 21

print("My name is {},I am {} years old.".format(name,age))    #1 way
print(f"My name is {name},I am {age} years old.")          #2 way
```

```python
a=7
b=2
print(f"Sum of {a} and {b} is {a + b}")
```

```python
# FORMATING NUMBERS
pi = 3.14159
print(f"round of {pi} to two decimal place {pi:.2f}")
```
