MATCH CASE:-

```python
def http_status(code):
    match code:
        case 200:
            return "ok"
        case 400:
            return "Bad request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown status"
```

```python
http_status(500)
```

```python
http_status(44)
```

```python
point = (3, 4)

match point:
    case (0, 0):
        print("Origin")
    case (x, 0):
        print(f"X-Axis at {x}")
    case (0, y):
        print(f"Y-Axis at {y}")
    case (x, y):
        print(f"Point at ({x}, {y})")
```
