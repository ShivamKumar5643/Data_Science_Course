```python
#Speed test of numpy and list
import numpy as np
import time 
```

```python
l1 = [1,2,4,5]    # this is zipping
l2 = [6,7,4,9]
list(zip(l1,l2))
```

```python
size = 1_000_000

list1 = list(range(size))
list2 = list(range(size))

arr1 = np.array(list1)
arr2 = np.array(list2)

# Python list addition
start = time.time()
result_list = [x + y for x, y in zip(list1, list2)]
end = time.time()
print("List time:", end - start)

# NumPy addition               #its backend written in c
start = time.time()
result_np = arr1 + arr2
end = time.time()
print("NumPy time:", end - start)
```
